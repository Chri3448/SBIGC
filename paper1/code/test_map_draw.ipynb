```python
%reset

import healpy as hp
import numpy as np
import time 

import matplotlib.pyplot as pl

%matplotlib inline
```

```python
nside = 2048
npix = hp.nside2npix(nside)
pix_size = np.sqrt(4.*np.pi/npix)*180./np.pi
print("pixel size (deg) = ", pix_size)
```

```python
frac_sky = 0.2
mask = np.arange(0,frac_sky*npix).astype('int')
print("len mask = ", len(mask))

the_map = np.zeros(npix)
t1 = time.time()
the_map[mask] = np.random.poisson(lam = 1., size = len(mask))
t2 = time.time()

#generate random 
n_photons = np.sum(the_map).astype('int')
uu = np.random.random(n_photons)
vv = np.random.random(n_photons)
theta = 2.*np.pi*uu
phi = np.arccos(2.*vv - 1.)

t3 = time.time()


print("t2 - t1 = ", t2-t1)
print("t3 - t2 = ", t3-t2)



hp.mollview(the_map)
```
